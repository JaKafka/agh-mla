```python
import lstm.utilities as lsutil

import torch
import torch.nn as nn
import numpy as np

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)
np.random.seed(42)

df = lsutil.load_dataset_from_file("./data/AAPL.csv")
X_train, y_train, X_test, y_test = lsutil.process_dataset_dataframe(df, test_size=0.05)

```

```python
import lstm.models as lsmodels

simple_regressor = lsmodels.SimpleLSTMRegressor().to(DEVICE)
training_data = lsutil.run_training_loop_RMSE(DEVICE, simple_regressor, 50, X_train, y_train, X_test, y_test)

lsutil.print_train_summary_RMSE(DEVICE, simple_regressor, training_data, X_train, y_train, X_test, y_test)
```

```python
larger_simple_regressor = lsmodels.SimpleLSTMRegressor(lstm_layers=2).to(DEVICE)
training_data = lsutil.run_training_loop_RMSE(DEVICE, larger_simple_regressor, 50, X_train, y_train, X_test, y_test)

lsutil.print_train_summary_RMSE(DEVICE, larger_simple_regressor, training_data, X_train, y_train, X_test, y_test)
```

```python
even_larger_simple_regressor = lsmodels.SimpleLSTMRegressor(lstm_layers=5).to(DEVICE)
training_data = lsutil.run_training_loop_RMSE(DEVICE, even_larger_simple_regressor, 50, X_train, y_train, X_test, y_test)

lsutil.print_train_summary_RMSE(DEVICE, even_larger_simple_regressor, training_data, X_train, y_train, X_test, y_test)
```

```python
even_larger_simple_regressor = lsmodels.SimpleLSTMRegressor(lstm_layers=5, hidden_size1=32).to(DEVICE)
training_data = lsutil.run_training_loop_RMSE(DEVICE, even_larger_simple_regressor, 50, X_train, y_train, X_test, y_test)

lsutil.print_train_summary_RMSE(DEVICE, even_larger_simple_regressor, training_data, X_train, y_train, X_test, y_test)
```
